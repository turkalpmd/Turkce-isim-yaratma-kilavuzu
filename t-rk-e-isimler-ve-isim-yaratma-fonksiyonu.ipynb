```python
import pandas as pd
import numpy as np
```

```python
df_soyisim = pd.read_csv("../input/turkce-isimler/soy_isimler.csv")
df_kadın = pd.read_csv("../input/turkce-isimler/kadn_isimleri.csv")
df_erkek = pd.read_csv("../input/turkce-isimler/erkek_isimleri.csv")
df_toplu = pd.read_csv("../input/turkce-isimler/karisik_isimler.csv")
df_soyisim.shape,df_kadın.shape,df_erkek.shape, df_toplu.shape
```

### Şimdi iki farklı durum olabilir, Elimizdeki verisetinde hali hazırda cinsiyetleri betimleyen bir öznitelik var veya yok.
#### Öznitelik olmadığını var sayarak bir veri seti uyduralım

```python
df_isim_1= pd.DataFrame()
df_isim_1["yaş"]= np.random.uniform(0,100,100).round(0)
df_isim_1["Z skor"] = np.random.normal(0,2,100)
df_isim_1["Kalsiyum_seviyesi"]= np.random.uniform(8,12,100).round(2)
df_isim_1["PTH"]= np.random.uniform(10,30,100).round(2)
random_cinsiyet= []
random_isim = []
        

## Daha sonra yine 100 kere isim ve soy isim datalarımızda gezinip random atamalar yapalım
for i in range(100):
    soyisim_ = df_soyisim["soyisim"].iloc[np.random.randint(0,1653)]
    tumisim_ = df_toplu["isimler"].iloc[np.random.randint(0,115)]
    isim_ = (tumisim_)+(" ")+(soyisim_)
    random_isim.append(isim_)
    
df_isim_1["random_isim"] = random_isim
df_isim_1
```

## İkinci bir ihtimal Datasette cinsiyet mevcut olabilir bu durumda

```python
df_isim_2= pd.DataFrame()
df_isim_2["yaş"]= np.random.uniform(0,100,100).round(0)
df_isim_2["Z skor"] = np.random.normal(0,2,100)
df_isim_2["Kalsiyum_seviyesi"]= np.random.uniform(8,12,100).round(2)
df_isim_2["PTH"]= np.random.uniform(10,30,100).round(2)
random_cinsiyet= []
random_isim = []


## Bu kısımların aynısı fonksiyonda da bulunacak
for i in range(df_isim_2.shape[0]):
    # Yani yüz kere for çalışsın 100 farklı cinsiyeti random yaratsın
    if np.random.randint(0,2) == 1:
        random_cinsiyet.append("Kadın")
    else:
        random_cinsiyet.append("Erkek")
df_isim_2["Cinsiyet"] = random_cinsiyet
df_isim_2
```

```python
# Daha gerçekçi olsun diye erkeğe erkek kadına kadın ismi atayabiliriz.
for i in range(df_isim_2.shape[0]):

    soyisim_ = df_soyisim["soyisim"].iloc[np.random.randint(0,1654)]
    kadın_ = df_kadın["kadın_isimleri"].iloc[np.random.randint(0,69)]
    erkek_ = df_erkek["erkek_isimleri"].iloc[np.random.randint(0,48)]

    if df_isim_2["Cinsiyet"].iloc[i] == "Kadın":
        isim_ = (kadın_)+(" ")+(soyisim_)
        random_isim.append(isim_)

    else:
        isim_ = (erkek_)+(" ")+(soyisim_)
        random_isim.append(isim_)
df_isim_2["random_isim"] = random_isim
df_isim_2
```
